##Enunciado

Sua tarefa ser  gerar a matriz termo documento, dos documentos recuperados da internet e 
imprimir esta matriz na tela. Para tanto: <br>
a) Considere que todas as listas de sentenas devem ser transformadas em listas de vetores, onde cada item ser uma das palavras da sentena.<br> 
b) Todos  os  vetores  devem  ser  unidos  em  um  corpus  nico  formando  uma  lista  de  vetores, <br>

##Imports

```python

import bs4
import requests
import re
import pandas

from google.colab import data_table
from google.colab import files

import ipywidgets as widgets
from IPython.display import display
data_table.enable_dataframe_formatter()
```

##Cdigo!



```python
def on_button_clicked(b):
  display(button, output)
  with output:
    for i in matriz:
      print(i)

def baixar(b):
    files.download("BagOfWords.xlsx")
```

```python
sites = ["https://www.ibm.com/cloud/learn/natural-language-processing", "https://en.wikipedia.org/wiki/Natural_language_processing", "https://www.techtarget.com/searchenterpriseai/definition/natural-language-processing-NLP", "https://www.sas.com/en_us/insights/analytics/what-is-natural-language-processing-nlp.html", "https://www.datarobot.com/blog/what-is-natural-language-processing-introduction-to-nlp/"]
conteudos = []
contagem = []
matriz = []
bagOfWords = []
palavras = []
#pegando os contedos dos sites
for site in sites:
    url = requests.get(site).content
    soup = bs4.BeautifulSoup(url, "html.parser")
    for data in soup(['style', 'script']):
        data.decompose()
    conteudo = ' '.join(soup.stripped_strings)
    conteudo = re.sub(r"[\n\t]", "", conteudo)
    splitters = re.split("[!?.;:,]", conteudo)
    conteudos.append(splitters)

#adicionando as palavras em um set para ter referncia de tudo
for conteudo in conteudos:
  for sentenca in conteudo:
    for palavra in sentenca.split():
      palavras.append(palavra)
palavras = set(palavras)
palavras = sorted(palavras)
#contando as palavras e adicionando as contagens na matriz
for conteudo in conteudos:
  contagem = [0] * len(palavras)
  for sentenca in conteudo:
    for palavra in sentenca.split():
      contagem[palavras.index(palavra)] += 1
  matriz.append(contagem)

#deixando tudo um pouco mais legvel e bonitinho
df = pandas.DataFrame(matriz, columns=palavras)
df.to_excel("BagOfWords.xlsx")


button = widgets.Button(description="Ver a matriz?")
output = widgets.Output()

button2 = widgets.Button(description="Download da Matriz?")
output2 = widgets.Output()

button.on_click(on_button_clicked)
button2.on_click(baixar)

```
